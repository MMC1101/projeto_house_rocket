```python
# libraries
# -----------------------------------------------------------------------------
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
import ploty.express as px

# Functions
# ------------------------------------------------------------------------------
df = data

def show_dtypes(df):
	t = df.dtypes
	return t

# # data dimensions


def see_the_shape(df):

	print(f'Number of rows: {df.shape[0]}')
	print(f'Number of columns: {df.shape[1]}')

	return None

# criando uma funçao que faz requisição API e preenche as novas colunas 

def collect_data(df, cols):
# initialize API

	geolocator = Nominatim = (user_agent = 'geoapiExercises')
		
	df = df.head(20)

		# create columns empy
		
		# for item range(len(cols)):
	df.loc[:, cols[0]] = 'NA'
	df.loc[:, cols[1]] = 'NA'
		
	for i in range(len(df)): 
			# percorre toda coluna lat e todas a coluna long, e concatena cada linha e de
		query = str(df.loc[i,'lat']) + ',' + str(df.loc[i,'long'])
		response = geolocator.reverse(query)
			# verifica se as colunas atribuidas no inicio da function existe no dicionario adress 
			# caso a coluna exista no dicionario preenche as colunas definidas 
		if cols[0] in response.raw['address']:
			df.loc[i, 'house_number'] = response.raw['address'][cols[0]]

		if cols[1] in response.raw['address']:
			df.loc[i, 'road'] = response.raw['address'][cols[1]]

	return df	
# --------------------------------------------------------------------------------
# extracao 
# -------------------------------------------------------------------------------
# load dataset-------------------------------------------------------------------
data = pd.read_csv('kc_house_data.csv')


# transformation
# --------------------------------------------------------------------------------
# descriptive statistics 

# central tendency - media , mediana
new_atribuites = data.select_dtypes(include = ['float64','int64'])

media = new_atribuites.apply(np.mean, axis = 0)

mediana = new_atribuites.apply(np.median, axis = 0)

# std, min , max 

min_ = new_atribuites.apply(np.min, axis= 0)
max_ = new_atribuites.apply(np.max, axis= 0)
std_ = new_atribuites.apply(np.std, axis = 0)

# transforming filters into dataframe
new_atribuites = pd.DataFrame(new_atribuites)
media = pd.DataFrame(media)
mediana = pd.DataFrame(mediana)
min_ = pd.DataFrame(min_)
max_ = pd.DataFrame(max_)
std_ = pd.DataFrame(std_)


df1 = pd.concat([min_, max_, media , mediana , std_], axis = 1 ).reset_index()

df1.columns= ['atribuites','minimo','maximo','media','mediana', 'std']

see_the_shape(df1)


# Crie uma nova coluna chamada: 'dormitory_type'
# - Se o valor da coluna 'bedrooms' for igual à 1 => 'studio'
# - Se o valor da coluna 'bedrooms' for igual à 2 => 'apartament'
# - Se o valor da coluna 'bedrooms' for maior que 2 => 'house'

data['dormitory_type'] = data['bedrooms'].apply(lambda x: 'studio' if x == 1 else 'apartament' if x == 2 else 'house'))

# criando uma nova coluna chamada data que classifica o nível dos imóveis de acordo com os prices

data['level'] = data['price'].apply(lambda x: 0 if x <= 321950 else 1 (x > 321950) & (x <= 450000) if (x > 450000) & (x <= 645000) else 3)


see_the_shape(data)



# # -------------------------------------------------------------------------------


# map 

houses = data[['id','lat', 'long', 'price', 'level']].copy()

fig = px.scatter_mapbox(houses, lat="lat", 
												lon="long",
												color = "level", 
												hover_name="id", 
												hover_data= "price" ,
                        color_discrete_sequence=["fuchsia"], 
												size_max = 15, 
												zoom=3)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
