```python
import pandas as pd
import numpy as np 

## importing project data 

data = pd.read_csv('kc_house_data.csv')
```

```python
## data dimensions 
def see_the_shape(df):
    
    print(f'Number of rows: {df.shape[0]}')
    print(f'Number of columns: {df.shape[1]}')

    return None
```

```python
see_the_shape(data)
```

```python
data.dtypes

### checking data typing
### converting data from date column to date
data['date'] = pd.to_datetime(data["date"])
```

```python
## descriptive statistics 

## central tendency - media , mediana
new_atribuites = data.select_dtypes(include = ['float64','int64'])

media = new_atribuites.apply(np.mean, axis = 0)

mediana = new_atribuites.apply(np.median, axis = 0)

## std, min , max 

min_ = new_atribuites.apply(np.min, axis= 0)
max_ = new_atribuites.apply(np.max, axis= 0)
std_ = new_atribuites.apply(np.std, axis = 0)
```

```python
## transforming filters into dataframe
new_atribuites = pd.DataFrame(new_atribuites)
media = pd.DataFrame(media)
mediana = pd.DataFrame(mediana)
min_ = pd.DataFrame(min_)
max_ = pd.DataFrame(max_)
std_ = pd.DataFrame(std_)


df1 = pd.concat([min_, max_, media , mediana , std_], axis = 1 ).reset_index()

df1.columns= ['atribuites','minimo','maximo','media','mediana', 'std']
```

```python
## removendo a notacao cientifica 

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```
